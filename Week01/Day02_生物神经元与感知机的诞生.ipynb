## 1. 灵感来源：生物神经元

人类大脑有大约 **860 亿个神经元**，它们通过电信号传递信息。一个神经元的工作流程是：

```text
  其他神经元的信号
       │
       ▼
  ┌─────────────────────┐
  │  树突 (接收信号)     │  ← 来自多个方向的输入
  │         │           │
  │    细胞体 (求和)     │  ← 把所有刺激加起来
  │         │           │
  │    超过阈值？        │  ← 判断：够不够强？
  │      /    \         │
  │    是      否       │
  │    ↓       ↓        │
  │  兴奋    沉默       │
  │  轴突 (输出信号)     │  ← 传递给下一个神经元
  └─────────────────────┘
```

关键观察：
- 一个神经元接收**多个输入**（来自其他神经元的信号）
- 每个输入的**重要程度不同**（有的信号强，有的弱）
- 只有总刺激**超过某个阈值**，才会「激活」并传递信号

罗森布拉特的天才之处：他发现这个生物过程，完全可以用**一个数学公式**来描述。

### 神经元 → 感知机的映射

| 生物神经元 | 数学感知机 |
|-----------|----------|
| 树突接收的多个信号 | 输入 X (一个向量) |
| 不同突触的连接强度 | 权重 W (每个输入对应一个权重) |
| 细胞体的求和 | 加权求和: X1*W1 + X2*W2 + ... |
| 神经元的兴奋倾向 | 偏置 Bias |
| 阈值判断 | 激活函数 |
| 轴突输出 | 最终输出 |

现在，让我们用代码把这个「骨架」搭建出来。

## 2. 搭建感知机骨架

我们采用「渐进式构建」的方法。今天先搭骨架（接收输入），明天再注入灵魂（权重和偏置）。

这样做是为了让你清楚地看到：感知机的每一部分能力是怎么一步步加上去的。

````python
import torch

class Perceptron:
    """
    感知机 v0.1 —— 最基础的骨架
    目前只能记录「我准备接收几个输入」，还不会做任何计算。
    """
    def __init__(self, input_features_count):
        self.input_features_count = input_features_count
        
        print(f"一个感知机诞生了！")
        print(f"  它可以同时接收 {self.input_features_count} 个输入信号")
        print(f"  但它现在还是个空壳，没有记忆，也不会思考...")

# 创建一个感知机
neuron = Perceptron(input_features_count=2)
````

### 类比一下

想象你刚买了一台电脑主机：
- `input_features_count=2` 意味着机箱上有 **2 个 USB 接口**
- 但里面没有 CPU、没有内存——它还不能运算

明天我们会给它装上「CPU」（权重）和「电源」（偏置）。

## 3. 多个感知机的组合：感知机层

单个感知机只能做一个二分类决策（是/否）。但在实际中，我们经常需要**多个感知机并行工作**，每个负责不同的判断。

比如判断「今天是否适合户外运动」，可能需要同时考虑：
- 感知机 A：天气怎么样？
- 感知机 B：有没有朋友一起？
- 感知机 C：身体状态如何？

这种「一堆感知机并排」的结构，就叫**感知机层**。

````python
class PerceptronLayer:
    """
    感知机层 —— 多个感知机并行工作
    """
    def __init__(self, input_count, output_count):
        self.input_count = input_count
        self.output_count = output_count
        
        # 创建 output_count 个感知机，每个都接收 input_count 个输入
        self.neurons = []
        for i in range(output_count):
            self.neurons.append(Perceptron(input_count))
        
        print(f"\n感知机层创建完成：{input_count} 个输入 -> {output_count} 个感知机并行工作")

# 3个输入信号，由2个感知机分别判断
layer = PerceptronLayer(input_count=3, output_count=2)
````

## 4. 历史背景：1957 年的狂欢

当罗森布拉特在 1957 年公布感知机时，引起了巨大的轰动。

《纽约时报》这样报道：
> 「海军今天展示了一台电子计算机的雏形，它期望海军能够有一天能够行走、交谈、看东西、写作、复制自己，并意识到自己的存在。」

当时的科学家们乐观地认为：只要把足够多的感知机堆叠起来，真正的智能就指日可待。

然而，1969 年，人工智能领域的权威马文·明斯基（Marvin Minsky）在《感知机》一书中，证明了一个致命的事实——

**单层感知机连最基本的异或（XOR）逻辑都无法表达。**

这个发现直接引发了 AI 的第一次寒冬，长达十余年无人问津。

我们将在第 4-5 天亲身体验这个「代码现场」。
